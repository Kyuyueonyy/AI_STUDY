# 데이터 분석

## 데이터 분석이란?
데이터 분석은 데이터를 수집, 정리, 처리하여 유용한 정보를 도출하고 의사결정을 지원하는 과정으로 데이터의 패턴, 추세, 관계를 파악하여 문제를 해결하거나 예측 모델을 구축하는 데 사용된다.

- 데이터는 숫자, 텍스트, 이미지 등 다양한 형태로 존재할 수 있다.
- 분석 결과는 문제 해결, 미래 예측, 또는 전략 수립에 활용된다.

### 데이터 분석의 주요 목적
1. 정보 도출: 데이터에서 의미 있는 인사이트를 발견.
2. 의사결정 지원: 데이터 기반으로 합리적인 결정을 내릴 수 있도록 도움.
3. 문제 해결: 데이터를 활용해 비즈니스 또는 사회적 문제를 해결.
4. 예측 및 모델링: 미래를 예측하거나 현상 이해를 위한 모델 생성.

### 데이터 분석의 주요 단계
1. 데이터 수집
  - 데이터를 다양한 소스에서 수집 (예: 설문조사, 로그 데이터, API 등).

2. 데이터 처리 및 정제
  - 결측치 처리, 이상치 제거, 데이터 변환 등을 통해 분석 가능한 상태로 정리.

3. 탐색적 데이터 분석(EDA)
  - 데이터의 구조, 분포, 패턴 등을 시각화하고 요약.

4. 데이터 분석 및 모델링
  - 통계 기법, 기계 학습, 데이터 마이닝을 활용해 데이터를 분석.
  - 예: 회귀 분석, 분류, 군집화 등.

5. 결과 시각화 및 해석
  - 시각화 도구를 통해 결과를 명확히 전달.
  - 예: 차트, 그래프, 대시보드 등을 활용.

6. 보고 및 의사결정
  - 분석 결과를 이해하기 쉽게 보고서로 작성하여 전략적 의사결정에 활용.

```python
from datasets import load_dataset
import pandas as pd
ds = load_dataset("bebechien/korean_cake_boss")

df = pd.DataFrame(ds["train"])
df.head()
```

### 2. 데이터 처리 및 정제
수집한 데이터를 분석 가능하도록 정리하고 정제하는 단계이다.

#### 주요 작업 내용:
  - 결측치 처리: 누락된 값을 대체하거나 제거.
  - 이상치 처리: 통계적 분석으로 이상값을 식별 및 조정.
  - 데이터 변환: 정규화, 표준화, 변수 인코딩 등.
  - 중복 제거: 중복 데이터 식별 및 제거.

#### 도구 및 기술:
- Python(Pandas, Numpy), SQL.
- 예시:
  - 잘못 입력된 고객 정보 수정 및 중복된 데이터 제거.
  - 온도 데이터를 섭씨로 변환.

```python
# "\n" 및 "\r" 제거
# regex=Flase 정규 표현식 지정 여부 | 정규 표현식으로 지정하게 되면 줄바꿈으로 처리된다.
df['input'] = df['input'].str.replace('\n', '', regex=False).str.replace('\r', '', regex=False)
df["output"] = df["output"].str.replace('\n', '', regex=False).str.replace('\r','',regex=False)

df.head()
```

### 3. 데이터 분석
정제된 데이터를 바탕으로 패턴, 관계, 추세 등을 탐구하여 인사이트를 도출.

#### 분석 유형:
  - 기술적 분석: 과거 데이터를 요약 (평균, 분산 등).
  - 진단적 분석: 이상 원인 분석 (상관 관계, 회귀 분석).
  - 예측 분석: 기계 학습을 통해 미래 예측.
  - 처방적 분석: 최적의 해결 방안 제시.

#### 도구 및 기술:
  - Python(Scikit-learn), R, Excel, Tableau.

#### 예시:
  - 고객 이탈률 분석 및 예측.
  - 판매 추세를 파악해 다음 분기의 매출 예측.

```python
import re

# 케이크 종류 구분
def extract_cake_type(text):
    # 정규 표현식 패턴
    pattern = r"\s*([\w\s]+케이크)|(\d+호\s*케이크)"
    matches = re.findall(pattern, text)  # 수정된 부분

    # 추출된 케이크 종류를 정리
    cake_types = [match[0] if match[0] else match[1] for match in matches]
    return [cake.strip() for cake in cake_types]

df["케이크_종류"] = df["output"].apply(extract_cake_type)

# 케이크 호수 구분
def extract_cake_size(text):
  pattern = r"(\d+호)"
  matches = re.findall(pattern, text)
  return matches[0] if matches else None

df["케이크_호수"] = df["input"].apply(extract_cake_size)
df[["케이크_종류", "케이크_호수"]]
```

### 4. 데이터 시각화

  - 시각적 스토리텔링으로 결과의 맥락 전달.

  - 대시보드를 통해 실시간 매출 데이터 제공.

```python
from IPython.display import display
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')

df['케이크_호수'] = df['케이크_호수'].astype(str).replace('None', 'None')

# 각 호수별로 개수 집계
cake_sales = df['케이크_호수'].value_counts().reset_index()
cake_sales.columns = ['케이크_호수', '판매_수']
display(cake_sales)

plt.figure(figsize=(10,6))
plt.bar(cake_sales["케이크_호수"], cake_sales["판매_수"], color="skyblue")
plt.title("호수별 판매량")
plt.xlabel("케이크 호수")
plt.ylabel("판매 수")
plt.show()
```
