## Bag-of-Words (BoW)
Bag-of-Workds는 문서 전체를 수치화하는 가장 기본적인 방법이다. <br>
BoW는 문서를 '단어들의 가방'으로 취급하여, 어떤 단어가 얼마나 나타나는지에 집중하는 방식이다.

> 동작 방식 <br>
> - 구현 방식: 문서를 마치 **'단어가 담긴 가방(Bag of Words)'**처럼 생각한다. <br> 가방 안에는 여러 단어들이 들어있지만, 이 단어들이 어떤 순서로 들어있는지는 전혀 중요하게 생각하지 않고, <br> 오직 각 단어가 얼마나 많이(빈도) 출현했는지에만 집중한다. <br>
> - 벡터화: 각 문서를 단어 사전(Vocabulary)에 있는 단어들의 출현 빈도로 구성된 벡터로 표현한다. <br> 이 벡터의 각 차원(feature)은 단어 사전의 특정 단어에 해당하고, 그 값은 해당 단어가 문서에 나타난 횟수가 된다.

```python
import re
import requests
from bs4 import BeautifulSoup


res = requests.get("https://kr.benzinga.com/news/usa/stocks/%EC%9C%A0%EB%82%98%EC%9D%B4%ED%8B%B0%EB%93%9C%ED%97%AC%EC%8A%A4-%EC%A3%BC%EC%8B%9D-11-%EA%B8%89%EB%93%B1-%EB%A7%88%EC%9D%B4%ED%81%B4-%EB%B2%84%EB%A6%AC%EC%99%80-%EC%9B%8C%EB%A0%8C-%EB%B2%84%ED%95%8F/")
soup = BeautifulSoup(res.text, 'lxml')
title = soup.find("h1").text


body = soup.find("div", class_="post-content cf entry-content content-spacious")

# 정규표현식을 사용하여 HTML 태그를 모두 찾아 빈 문자열로 대체
cleaned_text = re.sub(r'<[^>]*>', '', body.text)

print(title)
# 결과 출력 (앞뒤 공백 제거)
print(cleaned_text.strip())
```

```python
from konlpy.tag import Okt

# 1. 형태소 분석기로 명사만 추출 (토큰화)
okt = Okt()
nouns = okt.nouns(cleaned_text)

# 2. 불용어 및 한 글자 단어 제거
stopwords = ['위', '이', '것', '주', '수', '등', '년', '그', '약'] # 분석에 불필요한 단어 추가
meaningful_nouns = [word for word in nouns if word not in stopwords and len(word) > 1] # 형태소의 단어 중 불용어에 포함되지 않으면서 길이가 1 이상인 단어.

print("의미있는 명사 추출 (상위 20개):")
print(meaningful_nouns[:20])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# CountVectorizer는 문자열 리스트를 입력으로 받으므로, 명사 리스트를 공백으로 연결된 문자열로 변환
document = [' '.join(meaningful_nouns)]

# 1. BoW 모델 생성 (단어장 구축 및 벡터화)
vectorizer = CountVectorizer()
DTM = vectorizer.fit_transform(document)
print(DTM.toarray())
# 2. 결과 확인
# vocab: CountVectorizer가 학습한 단어 목록 (DTM의 열 이름)
vocab = vectorizer.get_feature_names_out()

# DTM: '희소 행렬(Sparse Matrix)' 형태의 문서-단어 행렬
# DTM.toarray(): 희소 행렬을 모든 0을 포함하는 '일반 배열(Dense Array)'로 변환
df_dtm = pd.DataFrame(DTM.toarray(), columns=vocab)

print("생성된 Bag-of-Words (DTM)")

# 단어가 너무 많으므로, 빈도수 상위 10개 단어만 확인
word_counts = df_dtm.sum().sort_values(ascending=False) # 내림차순 정렬
display(word_counts.head().to_frame(name='빈도수'))
```

## TF-IDF (Term Frequency-Inverse Document Frequency)
BoW는 문서에서 단어의 출현 빈도만 확인을 했지만, 단어의 등장 빈도가 높다고 중요한 것은 아니다. <br>
이를 해결하기 위해 TF-IDF는 단어가 문서에 얼마나 자주 등장하는지(TF)와 함께, 그 단어가  <br>
전체 문서 집합에서 얼마나 희귀한지(IDF)를 동시에 고려하여 단어의 가중치를 부여하는 모델이다.<br>
이를 통해 특정 단어가 어떤 문서를 대표하는지 알 수 있게된다.

> 핵심 개념
> - 단어 빈도 (TF, Term Frequency)
>  - 개념 : 특정 문서 내에서 단어가 얼마나 자주 등장하는지 확인
>  - $$ \text{TF}(\text{t}, \text{d}) = \frac{\text{특정 문서 d에서 단어 t의 등장 횟수}}{문서 d의 전체 단어 수}$$
> - 역문서 빈도 :  IDF (Inverse Document Frequency)
>  - 개념 : 특정 단어가 전체 문서에 얼마나 분포되어 있는지를 반영한다.
>  -  $$ \text{IDF}(t) = \log\left(\frac{전체 문서 수}{\text{특정 단어 t가 등장한 문서의 수}}\right)$$
> - 계산식
>  - $$ \text{TF-IDF}(t, d) = \text{TF}(t, d) \times \log\left(\frac{N}{\text{DF}(t)}\right)$$
> - 키워드
  - t : 단어
  - d : 문서
  - n : 전체 문서 수
  - TF-IDF(t,d) : 문서에서 단어의 출현 빈도
  - DF(t) : 단어(t)가 등장하는 문서수

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 올바른 입력 데이터 형태 (각 리스트가 하나의 문서를 의미)
# meaningful_nouns는 문서들의 리스트여야 함
meaningful_nouns = [
    ['유나이티드헬스', '보험', '가입', '방법'],
    ['건강보험', '혜택', '비교'],
    ['유나이티드헬스', '병원', '이용', '안내']
]

# 1. TfidfVectorizer 객체 생성 및 TF-IDF 행렬 구축
# 입력 데이터를 공백으로 연결된 문자열 리스트로 변환
documents = [' '.join(doc) for doc in meaningful_nouns]
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# 2. 결과 확인
vocab = tfidf_vectorizer.get_feature_names_out()
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vocab)

print("## 생성된 TF-IDF 행렬 ##")
print(df_tfidf)

print("\n--- '유나이티드헬스'의 점수 ---")
print(df_tfidf['유나이티드헬스'])
```
