### 1. Beautiful Soupì˜ í•µì‹¬ ê°ì²´ ğŸŒ³
Beautiful Soupì€ HTML ë¬¸ì„œë¥¼ ëª‡ ê°€ì§€ ì¢…ë¥˜ì˜ ê°ì²´ë¡œ ë‚˜ëˆ„ì–´ ë‹¤ë£¬ë‹¤. <br>

- BeautifulSoup: ì „ì²´ HTML ë¬¸ì„œë¥¼ ì˜ë¯¸í•˜ëŠ” 'ê°€ë¬¸ ì „ì²´' ì™€ ê°™ë‹¤. ìš°ë¦¬ê°€ soup ë³€ìˆ˜ì— ì €ì¥í•œ ìµœìƒìœ„ ê°ì²´ì´ë‹¤.

- Tag: `<h1>`, `<p>`, `<a>` ë“± ê°ê°ì˜ HTML íƒœê·¸ë¥¼ ì˜ë¯¸í•˜ëŠ” 'ê°€ì¡± êµ¬ì„±ì›' ì´ë‹¤. íƒœê·¸ëŠ” ë˜ ë‹¤ë¥¸ íƒœê·¸ë¥¼ ìì‹ìœ¼ë¡œ ê°€ì§ˆ ìˆ˜ ìˆë‹¤.

- NavigableString: íƒœê·¸ ì•ˆì— ë“¤ì–´ìˆëŠ” í…ìŠ¤íŠ¸(ë¬¸ìì—´)ì´ë‹¤. ê°€ì¡± êµ¬ì„±ì›ì˜ 'ì´ë¦„' ì´ë‚˜ 'ëª©ì†Œë¦¬' ì— í•´ë‹¹í•œë‹¤.

- Comment: `` ë¶€ë¶„ìœ¼ë¡œ, í”„ë¡œê·¸ë˜ë¨¸ì—ê²Œë§Œ ë³´ì´ëŠ” ë©”ëª¨ì´ë‹¤.

```python
from bs4 import BeautifulSoup

# html_doc <- ê°€ì¥ ì²˜ìŒ ì¡°íšŒí•œ ë³€ìˆ˜
soup = BeautifulSoup(html_doc, 'lxml')

tag_h1 = soup.h1 # h1 íƒœê·¸ ê°ì²´ (Tag)ë¡œ ê°€ì¥ ë¨¼ì € ë°œê²¬ë˜ëŠ” `h1`ì„ ë°˜í™˜
print(f"h1 íƒœê·¸ ê°ì²´: {tag_h1}, íƒ€ì…: {type(tag_h1)}")

text_in_a = tag_h1.a.string # a íƒœê·¸ ì•ˆì˜ í…ìŠ¤íŠ¸ (NavigableString)
print(f"a íƒœê·¸ ì•ˆì˜ í…ìŠ¤íŠ¸: '{text_in_a}', íƒ€ì…: {type(text_in_a)}")

comment = soup.body.contents[2] # ì£¼ì„ ê°ì²´ (Comment)
print(f"ì£¼ì„ ê°ì²´: {comment}, íƒ€ì…: {type(comment)}")
```

### 2. HTML íŠ¸ë¦¬ íƒìƒ‰ (Navigating the Tree)
`find()`ë‚˜ `select()` ì—†ì´ë„, íŠ¹ì • íƒœê·¸ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶€ëª¨-ìì‹-í˜•ì œ ê´€ê³„ë¥¼ ì´ìš©í•´ íŠ¸ë¦¬ë¥¼ ììœ ë¡­ê²Œ ì´ë™í•  ìˆ˜ ìˆë‹¤.

- find : ëŒ€ìƒ íƒœê·¸ì—ì„œ ê°€ì¥ ë¨¼ì € ì°¾ì€ ë…¸ë“œë¥¼ ë°˜í™˜í•œë‹¤.
- find_all : ì°¾ìœ¼ë ¤ëŠ” íƒœê·¸ë¥¼ ëª¨ë‘ ì¡°íšŒí•œë‹¤.
- node.next_sibling : í˜„ì¬ íƒœê·¸ì˜ ë°”ë¡œ ë‹¤ìŒ í˜•ì œ ë…¸ë“œë¥¼ ë°˜í™˜í•œë‹¤.
  - íƒœê·¸ ì‚¬ì´ì˜ **ê³µë°±ì´ë‚˜ ì¤„ë°”ê¿ˆ ë¬¸ì(NavigableString)**ë„ ë°˜í™˜í•  ìˆ˜ ìˆë‹¤.
- find_next_sibling : íƒœê·¸ê°€ ì•„ë‹Œ ë¬¸ìì—´ì„ ê±´ë„ˆë›°ê³  ë‹¤ìŒ íƒœê·¸ í˜•ì œë¥¼ ì§ì ‘ ì°¾ëŠ”ë‹¤.
- node.parent : ëŒ€ìƒ ì»¬ëŸ¼ì˜ ë¶€ëª¨ ë…¸ë“œë¥¼ ì°¾ëŠ”ë‹¤.

```python
from bs4 import BeautifulSoup

html_family = """
<html>
  <body>
    <div class="family">
      <p id="father">ì•„ë¹ 
        <b id="son">ì•„ë“¤</b>
        <i id="daughter">ë”¸</i>
      </p>
      <p id="uncle">ì‚¼ì´Œ</p>
    </div>
  </body>
</html>
"""
soup = BeautifulSoup(html_family, 'lxml')

son = soup.find(id="son") # 'ì•„ë“¤'ì„ ê¸°ì¤€ìœ¼ë¡œ ì‹œì‘

# ë¶€ëª¨ ë…¸ë“œë¡œ ì´ë™í•˜ê¸°
father = son.parent
print(f"ì•„ë“¤ì˜ ë¶€ëª¨(.parent): {father.name} íƒœê·¸, idëŠ” {father['id']}")

# .next_siblingìœ¼ë¡œ í˜•ì œ ë…¸ë“œ ì°¾ê¸°
# 'ì•„ë“¤' íƒœê·¸ì™€ 'ë”¸' íƒœê·¸ ì‚¬ì´ì˜ ì¤„ë°”ê¿ˆ ë¬¸ì(NavigableString)ê°€ ë¨¼ì € ë°˜í™˜ë¨
daughter_via_sibling = son.next_sibling
print(f"son.next_siblingì˜ ê²°ê³¼ (ê³µë°± ë¬¸ìì—´): '{daughter_via_sibling}'")

daughter = daughter_via_sibling.next_sibling
print(f"ì•„ë“¤ì˜ ë‹¤ìŒ í˜•ì œ(.next_sibling.next_sibling): {daughter.name} íƒœê·¸, idëŠ” {daughter['id']}")

# .find_next_sibling()ìœ¼ë¡œ í˜•ì œ íƒœê·¸ ë°”ë¡œ ì°¾ê¸°
# ê³µë°± ë¬¸ìë¥¼ ê±´ë„ˆë›°ê³  ë‹¤ìŒ 'íƒœê·¸'ë¥¼ ì§ì ‘ ì°¾ëŠ”ë‹¤.
daughter_direct = son.find_next_sibling()
print(f"ì•„ë“¤ì˜ ë‹¤ìŒ í˜•ì œ(.find_next_sibling): {daughter_direct.name} íƒœê·¸, idëŠ” {daughter_direct['id']}")

all_p = soup.find_all("p")
for p in all_p :
  print(p)
```

## í•µì‹¬ ë¬¸ë²• II: CSS ì„ íƒìì™€ ë°ì´í„° ì¡°ì‘ (Advanced Implementation Layer)

### 1. CSS ì„ íƒì(Selector)ë¡œ ê²€ìƒ‰í•˜ê¸° ğŸ¯
ì›¹ ë¸Œë¼ìš°ì €ì˜ ê°œë°œì ë„êµ¬(F12)ì—ì„œ ìš”ì†Œë¥¼ ì„ íƒí•˜ê³  'Copy > Copy selector'ë¥¼ í•˜ë©´ ì–»ì„ ìˆ˜ ìˆëŠ” ê²½ë¡œê°€ ë°”ë¡œ CSS ì„ íƒìë‹¤. ì´ ë¬¸ë²•ì„ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•  ìˆ˜ ìˆë‹¤.

- select_one('ì„ íƒì'): ì„ íƒìì— í•´ë‹¹í•˜ëŠ” ì²« ë²ˆì§¸ ìš”ì†Œ í•˜ë‚˜ë¥¼ ë°˜í™˜í•œë‹¤. (find()ì™€ ìœ ì‚¬)

- select('ì„ íƒì'): ì„ íƒìì— í•´ë‹¹í•˜ëŠ” ëª¨ë“  ìš”ì†Œë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜í•œë‹¤. (find_all()ê³¼ ìœ ì‚¬)

```python
from bs4 import BeautifulSoup

html_shop = """
<html><body>
<div id="itemList">
  <div class="item">
    <a href="/product/1" class="name">ë…¸íŠ¸ë¶</a>
    <p class="price">1,200,000ì›</p>
  </div>
  <div class="item special">
    <a href="/product/2" class="name">ë§ˆìš°ìŠ¤</a>
    <p>ê°€ê²© ë¬¸ì˜</p>
  </div>
</div>
</body></html>
"""
soup = BeautifulSoup(html_shop, 'lxml')

# íƒœê·¸ ì„ íƒì
p_tag = soup.select_one('p')
print(f"íƒœê·¸ ì„ íƒì('p'): {p_tag.get_text()}")

# í´ë˜ìŠ¤ ì„ íƒì (.í´ë˜ìŠ¤ëª…)
item = soup.select_one('.item')
print(f"í´ë˜ìŠ¤ ì„ íƒì('.item'): {item.a.get_text()}")

# ID ì„ íƒì (#IDëª…)
item_list = soup.select_one('#itemList')
print(f"ID ì„ íƒì('#itemList'): {len(item_list.select('.item'))}ê°œì˜ ìƒí’ˆì´ ìˆìŒ")

# ìì† ê²°í•©ì (ìƒìœ„ í•˜ìœ„): ë„ì–´ì“°ê¸°ë¡œ êµ¬ë¶„
# '#itemList' ì•ˆì— ìˆëŠ” ëª¨ë“  'a' íƒœê·¸
all_names = soup.select('#itemList a')
print(f"ìì† ê²°í•©ì('#itemList a'): {[name.get_text() for name in all_names]}")

# ìì‹ ê²°í•©ì (ìƒìœ„ > ë°”ë¡œ ì•„ë˜ í•˜ìœ„): > ë¡œ êµ¬ë¶„
# '.item' ë°”ë¡œ ì•„ë˜ì— ìˆëŠ” 'a' íƒœê·¸
direct_child_name = soup.select_one('.item > a')
print(f"ìì‹ ê²°í•©ì('.item > a'): {direct_child_name.get_text()}")

# ì†ì„± ì„ íƒì ([ì†ì„±="ê°’"])
product_link = soup.select_one('a[href="/product/1"]')
print(f"ì†ì„± ì„ íƒì('a[href=\"/product/1\"]'): {product_link.get_text()}")
```

### 3. HTML íŠ¸ë¦¬ ìˆ˜ì •í•˜ê¸° (Modifying the Tree)
