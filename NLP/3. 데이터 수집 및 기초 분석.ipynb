# 데이터 수집

## 데이터 수집 방법론
데이터 수집 방법론은 분석 또는 의사결정을 위해 데이터를 체계적으로 모으는 과정으로, 데이터의 출처와 성격에 따라 1차 데이터와 2차 데이터로 구분된다.

### 1. 1차 데이터(Primary Data)
1차 데이터는 연구나 분석 목적으로 직접 수집한 데이터로, 특정 문제나 목표에 맞춰 설계된다.

- 특징
  - 맞춤형: 특정 목적을 위해 설계 및 수집됨.
  - 최신성: 데이터를 직접 생성하므로 최신 데이터를 확보 가능.
  - 독점성: 수집한 조직이나 개인만 접근 가능.
- 장점
  - 목적에 부합하는 데이터 확보.
  - 데이터의 출처와 품질에 대한 신뢰성 보장.
  - 특정 요구 사항에 따라 유연하게 설계 가능.
- 단점
  - 비용과 시간이 많이 소요됨.
  - 대규모 데이터 수집이 어려울 수 있음.
  - 데이터 수집 과정에서 오류 발생 가능성.

###  2. 2차 데이터 (Secondary Data)
2차 데이터는 다른 목적을 위해 이미 수집되고 저장된 데이터를 활용하는 것으로, 기존 데이터 소스를 활용한다.

- 특징
  - 기성 데이터: 다른 기관, 조직 또는 연구자가 수집한 데이터.
  - 경제적: 추가 비용 없이 기존 데이터를 재사용 가능.
  - 포괄성: 대규모 또는 장기간의 데이터를 포함할 수 있음.
- 장점
  - 시간과 비용 절약.
  - 대규모 데이터셋 확보 가능.
  - 다양한 출처를 통해 넓은 시야 제공.
- 단점
  - 데이터 품질과 신뢰성 문제.
  - 필요한 데이터가 목적에 맞지 않을 가능성.
  - 최신 데이터가 아닐 수 있음.

#### 수집 방법
1. 공공 데이터(Public Data):
  - 정부, 공공기관에서 제공하는 데이터.
  - 예: 국가 통계, 오픈 데이터 포털.
    1. [공공 데이터 포털](https://www.data.go.kr/index.do)
    2. [kaggle](https://www.kaggle.com/)
    3. [hugging Face](https://huggingface.co/?activityType=update-dataset&feedType=following)
    4. [AI-hub](https://www.aihub.or.kr/)

2. 기업 데이터(Corporate Data):
  - 민간 기업에서 생성한 데이터.
  - 예: 소셜 미디어 데이터, 구매 내역.
    1. [Facebook Graph API](https://developers.facebook.com/docs/graph-api/)
    2. [Reddit API](https://www.reddit.com/dev/api/)

3. 문헌 연구(Literature Review):
  - 학술 논문, 연구 보고서 등에서 데이터 추출.
  - 예: 경제 연구 보고서의 수치 데이터.

4. 웹 데이터(Web Scraping):
  - 웹사이트에서 데이터를 추출.
  - 예: 전자상거래 사이트의 상품 리뷰, 가격 데이터.

5. 데이터베이스 및 API:
  - 클라우드 기반 데이터 서비스나 API를 통해 데이터 제공.
  - 예: Google Analytics, Twitter API.


| 구분         | 1차 데이터                           | 2차 데이터                         |
|--------------|-------------------------------------|------------------------------------|

### 1. 문제 정의 (Problem Definition)
- **가장 큰 문제:** 옥지가 만족할 만한 영화를 찾아야 한다.

### 2. 무엇을, 어디서 찾을 것인가? (What & Where to Find)
- **무엇을(What):** 공식 '장르' 정보(예: 스릴러, 드라마)만으로는 옥지의 복합적인 감정 요구('슬프면서 재미있는')를 만족시킬 수 없다. 실제 영화를 본 사람들의 생생한 감상평, 즉 **'리뷰 데이터'**가 반드시 필요하다. 리뷰 안에 '감동적이었다', '스릴 넘친다' 같은 키워드가 있을 것이기 때문이다.
- **어디서(Where):** 이런 리뷰는 어디에 가장 많을까? → 사람들이 가장 많이 사용하는 **네이버, 다음** 같은 포털 사이트의 영화 정보 섹션.

### 3. 어떻게 해결할 것인가? (How to Solve)
- **해결 계획 수립:**

    3.  **(C) 마지막으로,** 수집된 리뷰 데이터에서 옥지가 말한 **감정 키워드('감동', '스릴' 등)가 얼마나 자주 등장하는지 분석**하여 최종 영화를 추천한다.

#### [1단계: 현재 상영작 정보를 얻기]


```python
import requests # 파이썬에서 api 요청을 위해 사용하는 lib

# 목표: 네이버 '현재 상영 영화' 검색 결과 페이지의 HTML 전체를 코드로 가져온다.
url = 'https://search.naver.com/search.naver?query=상영 영화'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'
}

# 빵빵이가 네이버에 검색하듯, 코드가 해당 URL에 접속해 페이지 전체 정보를 '요청'하고 '응답'받는다.
response = requests.get(url, headers=headers)

print(response.text)
```

> 장점:
> - 빠른 정보 획득: 브라우저를 켜고, 검색어를 입력하고, 결과를 기다리는 모든 과정을 단 코드 몇 줄로 순식간에 끝냈다.

#### [2단계 : BeautifulSoup을 이용한 html 파싱]

> BeatifulSoup
> - Beautiful Soup은 HTML 및 XML 문서에서 원하는 데이터를 쉽게 추출하도록 돕는 파이썬 라이브러리이다.
> - 웹 스크레이핑(Web Scraping) 과정에서 requests 라이브러리와 함께 거의 항상 사용되는 핵심 도구이다.


- html은 특정 구조를 가지고 있는데 이것을 쉽게 분석할 수 있도록 해주는 것이 BeatifulSoup이다. <br>
이 생각을 가지고 아래의 내용을 고민해야한다.



> [컴퓨터에게 일시키기]

  - 컴퓨터에게 일을 시키기 전에, 사람이라면 어떻게 영화 제목을 찾을지 생각해 보자.
  1. 웹 브라우저에서 '개발자 도구(F12)'를 켭니다.
  2. '요소 선택(Ctrl+Shift+C)' 아이콘을 클릭하고, 웹페이지의 영화 제목 위로 마우스를 가져간다.
  3. 개발자 도구 창에 해당 제목을 감싸고 있는 HTML 태그가 하이라이트되는 것을 본다. <br> <div class="data_box" ...> 와 같은 '주소'를 발견할 수 있다.
  4. 우리의 목표는 파이썬 코드가 이 '주소'를 찾아가서 그 안에 있는 텍스트를 가져오게 하는 것이다.

```python
from bs4 import BeautifulSoup # BeautifulSoup 라이브러리 추가
import time
movies = []
base_link = "https://search.naver.com/search.naver"

# 1단계에서 얻은 response.text를 BeautifulSoup으로 분석(해부) 가능한 객체로 변환
soup = BeautifulSoup(response.text, 'html.parser')
# [사고하기]에서 찾아낸 '주소(선택자)'를 이용해 모든 영화 제목 태그를 리스트로 가져온다.
movie_tags = soup.find_all('div', class_='card_item')

# 찾은 태그들(리스트)을 반복문으로 돌면서, 각 태그 안의 텍스트(.text)만 추출하여 출력
print("--- 현재 상영 영화 목록 ---")
for i, tag in enumerate(movie_tags):
    title_tag = tag.select_one('a.this_text')

      # 2. <a> 태그가 존재하는지 확인한다 (매우 중요).
    if title_tag:
      # 3. <a> 태그 안에서 제목(.text)과 링크(['href'])를 모두 추출한다.
      movie_title = title_tag.text
      movie_link = title_tag['href'] # <a> 태그에서 href를 가져온다.

      movie = {
          "no" : i+1,
          "title" :  movie_title,
          "movie_link" : base_link+movie_link
      }
      movies.append(movie)
      time.sleep(1) # 1초 대기 너무 빠르게 조회하면 요청을 막는다.

for content in movies:
  print(content)
```

> 위 목록은 현재 영화관에서 상영하는 영화이다.

```python
print("--- 영화별 리뷰 수집하기 ---")

for i,content in enumerate(movies):
  response = requests.get(content["movie_link"] + "평점")
  soup_response = BeautifulSoup(response.text, "html.parser")
  del movies[i]["movie_link"]

  avg = soup_response.find("span", class_="area_star_number")
  score = avg.text[:-2]
  movies[i]["avg"] = score

  reviews = []
  res_reviews = soup_response.find_all("span", class_="desc _text")
  for text in res_reviews:
    reviews.append(text.text)

  movies[i]["reviews"] = reviews

for content in movies:
  print(content)
```

```python
import pandas as pd

# 1. 'movies' 리스트 전체를 DataFrame으로 한 번에 변환
df = pd.DataFrame(movies)

# 2. 'reviews' 컬럼을 기준으로 explode() 실행
reviews_df = df.explode('reviews').reset_index(drop=True)

# 3. 결과 확인
display(reviews_df)
```

> - (C) 각 영화의 리뷰 목록(revies)을 하나씩 읽어본다.
> - (D) 각 리뷰 안에 (A)에서 정한 키워드가 들어있는지 확인한다.
> - (E) 한 영화의 전체 리뷰에서 4개의 키워드가 모두 한 번 이상 발견되었다면, 그 영화가 바로 빵빵이가 찾아 헤매던 정답이다!

```python
# 'reviews_df'는 이전 단계에서 explode로 만든 DataFrame입니다.

# 1. 옥지의 요구사항을 키워드 리스트로 정의
search_keywords = ["감동","로맨스", "스릴", "슬픔", "재미", "웃음"]

# 2. 각 키워드가 리뷰에 포함되어 있는지 확인하는 새로운 컬럼(플래그)들을 만듭니다.
for keyword in search_keywords:
    # .str.contains()는 텍스트 포함 여부를 True/False로 알려줍니다.
    # na=False는 리뷰가 없는 경우(NaN)를 False로 처리해 오류를 방지합니다.
    reviews_df[keyword] = reviews_df['reviews'].str.contains(keyword, na=False)

# 3. 영화 제목('title')별로 그룹화하여, 각 키워드가 한 번이라도 등장했는지 집계.
# .any()는 그룹 내에 True가 하나라도 있으면 결과가 True.
movie_analysis = reviews_df.groupby('title')[search_keywords].any()

# 4. 모든 키워드 조건(4개의 컬럼이 모두 True)을 만족하는 영화를 찾습니다.
# .all(axis=1)은 해당 행의 모든 값이 True일 때만 True를 반환합니다.
recommended_movies = movie_analysis[movie_analysis.all(axis=1)]


print("--- 영화별 키워드 분석 결과 ---")
# 각 영화가 어떤 키워드를 포함하는지 전체 요약표를 보여줍니다.
display(movie_analysis)


print("\n\n--- 빵빵이의 최종 추천 ---")
if not recommended_movies.empty:
    print("🎉 옥지의 모든 조건을 만족하는 영화를 찾았습니다!")
    # .index.tolist()는 조건을 만족하는 영화 제목들을 리스트로 보여줍니다.
    print(recommended_movies.index.tolist())
else:
    print("😥 아쉽게도, 현재 데이터로는 옥지의 모든 조건을 만족하는 영화는 없었습니다...")
```

## 빵빵이의 마지막 깨달음, 그리고 인공지능의 필요성
빵빵이는 파이썬을 이용해 데이터를 수집하고, 정제하고, 분석하여 마침내 옥지의 까다로운 질문에 답할 수 있는 수준까지 도달했지만 그는 코딩으로 문제를 접근할 수 있다는 것에 뿌듯해했다.

하지만 빵빵이는 곧 자신의 분석에 치명적인 허점이 있다는 것을 깨닫게 된다.

1. 단순 키워드 분석의 명백한 한계
  - 우리가 사용한 .str.contains() 방식은 빠르고 간단하지만, 인간 언어의 복잡성을 전혀 이해하지 못한다.
  - 즉, 특정 키워드만 추출함.

2. 유의어 문제 (The Synonym Problem)
  - 우리는 "감동"이라는 단어만 찾았지만, 만약 리뷰에 "가슴 뭉클한 영화였어요", "눈물이 났어요", "감동적인 서사" 라고 쓰여있었다면 어땠을까? 사람은 모두 같은 의미로 이해하지만, 우리의 코드는 이 소중한 리뷰들을 전부 놓치고 만다.
  - 즉, 화용론과 같이 돌려서 표현하는 경우 의미를 파악하지 못한다.

3. 부정어 문제 (The Negation Problem)
  - 만약 리뷰에 "솔직히 재미는 별로 없었다" 또는 "감동을 기대했는데 실망했다" 라는 문장이 있었다면? 우리의 코드는 "재미", "감동"이라는 단어가 포함되었다는 이유만으로 이 리뷰를 긍정적인 신호로 잘못 판단하게 된다.

4. 맥락 및 뉘앙스 문제 (The Context & Nuance Problem)
  - 인간의 언어는 단순히 단어의 나열이 아닙다. "아이와 보기엔 좋지만, 어른들이 보기엔 유치하다" 와 같은 문장은 긍정과 부정이 섞여있다.
  - 단순 키워드 검색으로는 이런 미묘한 맥락을 전혀 파악할 수 없다.

### 인공지능이 필요한 이유
이처럼 인간 언어가 가진 **'맥락'과 '의미'**의 벽에 부딪혔을 때, 단순한 코딩 규칙만으로는 정확한 분석에 한계가 명확해진다. <BR>
바로 이 지점에서 인공지능(AI), 특히 자연어 처리(NLP) 기술이 필요해지는 것이다.

> 의미 기반 분석
> - 인공지능은 "감동", "뭉클함", "눈물"이라는 단어들이 서로 의미적으로 가깝다는 것을 학습한다.
> - 단순히 글자가 같은지만 보는 게 아니라, 단어의 의미 자체를 이해하고 비슷한 단어를 함께 고려할 수 있다.
>
> 감성 분석 (Sentiment Analysis)
> - 인공지능은 수많은 문장을 학습하여 "재미없다"는 부정적인 감정을, "재미있다"는 긍정적인 감정을 담고 있다는 것을 안다.
> - 문장 전체의 구조와 뉘앙스를 파악하여 글의 진짜 속마음(긍정/부정)을 판단한다.
>
> 문맥 파악
> - "A를 기대했지만 B는 아쉬웠다"와 같은 복잡한 문장 구조를 이해하고, 어떤 부분에 대해 긍정적인지, 어떤 부분에 대해 부정적인지 분리해서 분석할 수 있다.
